## information about data set

The Cigarette data is a data frame consisting of 528 cases, each displaying information across 9 columns.

These columns are:
   
    1. state 
        consisting of 48 states (categorical w/48 levels).
    2. year
        integer ranging from 1985 to 1995 (year is a discretized measure of a continuous interval variable, so quantitative. Year can also be an ordinal variable (categorical)).
    3. cpi
        number representing changes in price level of a weighted average market basket of goods, in this case cigarettes (number that is quantitative or categorical).
    4. pop
        population is the given case resident count for the corresponding year (integer that is quantitative).
    5. packpc
        packs per capita is a number representing a states packs consumed per capita for the given year (quantitative).
    6. income
        integer of that state's income for the year (quantitative or categorical).
    7. tax 
        number detailing taxes of states by year (quatitative or categorical).
    8. avgprs
        number detailing the average price (quantitative or categorical).
    9. taxs
        another number detailing taxes (quantitative or cetegorical).

## Examining the packs per capita in 1995 vs the packs per capita in 1985

Create two data frames, one for each year in question. Then get a vector of packpc. Finally, we will use a paired t-test to see whether the number of packs per capita in 1995 was significantly different than the number of packs per capita in 1985

```r
cig85 = Cigarette %>% filter(year == 1985)
```

```r
cig95 = Cigarette %>% filter(year == 1995)
```

```r
head(cig85)
```

```r
packs85 = cig85$packpc
```

```r
packs95 = cig95$packpc
```

```r
head(packs85)
```

### paired t-test time

* Null hypothesis: There is no difference in means of packs per capita between 1985 and 1995.
* Alternative hypothesis: There is a difference in the means. The means are not equal. 

```r
t.obj = t.test(packs95, packs85, paired = TRUE)
```

```r
t.obj
```

Right off the bat, we can see that there a difference of 25.70863 between the means. The p-value is significant. Reject the null, accept the alternative.

Let's get the means and compare side by side.

```r
median(cig85$packpc)
```

```r
median(cig95$packpc)
```

Packs per capita in 1985 seems to have been a lot more than what was recorded ten years later!

```r
plotNormalHistogram(cig85$packpc)
## visual of where mean of 1985 packs per capita lies
```

```r
plotNormalHistogram(cig95$packpc)
## visual of where mean of 1995 packs per capita lies
```

## Basic ANOVA to determine whether a state's population influences its packs per capita

Start with recoding population into a categorical variable to compare it to packpc, which is continuous

```r
quantile(Cigarette$pop)
```

```r
CigData = Cigarette
```

```r
CigData$popClass <- cut(CigData$pop, breaks = c(0, 1579248, 3628253.5, 6012163, 31493524),
                  labels = c("small", "medSmall", "medLarge", "Large"))
```

```r
unique(CigData$popClass)
```

```r
head(CigData)
```

```r
CigData %>% group_by(popClass) %>% summarise(medvariable = median(pop))
```

### Testing assumptions

```r
## verify our continuous variable of packpc is normally distributed
```

```r
plotNormalHistogram(CigData$packpc)
## maybe a little positively skewed
## will try a correction with a sqrt
```

```r
CigData$packpcSQRT = sqrt(CigData$packpc)
```

```r
head(CigData)
```

```r
plotNormalHistogram(CigData$packpcSQRT)
## looks a lot more aproximately normally distributed
```

With parametric data, we use a Bartlett's Test to test for homogeneity of variance.

```r
bartlett.test(packpcSQRT ~ popClass, data = CigData)
```

With a p-value of .02394, the data has violated the assumption of homogeneity of variance.

```r
str(CigData)
## we have 528 cases in our sample size
```

### ANOVA when data has heterogeneity of variance

```r
cigANOVA = aov(CigData$packpcSQRT ~ CigData$popClass)
```

```r
summary(cigANOVA)
```

**With a p-value significant at .001, we can wee that a state's population class influence over its packs per capita.**

### Post Hocs

```r
## must account for violation of assumption of homogeneity of variance
pairwise.t.test(CigData$packpc, CigData$popClass, p.adjust = 'bonferroni', pool.sd = FALSE)
```

**The pairwise t-test shows significant differences amongst several categories. Small to medSmall, medSmall to medLarge, and medLarge to Large all show significance through their corresponding p-values.**

Time to determine the means and draw conclusions

```r
cigMeans = CigData %>% group_by(popClass) %>% summarize(Mean = mean(packpc))
```

```r
cigMeans
```

* _States with a small population class (< 1.579 million resident count) have an average of 110.6 packs per capita._
* _States in the medium small range (small < , < 3.628 million population) have an average of 98.8 packs per capita._
* _Medium large states (medSmall <, < 6.012 million) have an average of 112.2 packs per capita._
* _Large states (medLarge <) have an average of 104 packs per capita._

## Modeling with step-wise regression

Will be looking at how any predictor variables influence Y.
Let's have our dependent variable be packs per capita, and see if we can develop a model that will best predict it!

There are three approaches to stepwise regression:

* _Backward Elimination_
* _Forward Elimination_
* _Hybrid_

### Backward Elimination

Starting with all predictors, begin eliminating 1 variable at a time. Start with the IV that has the least effect on improving the model. Keep removing terms until the point at which removing a variable has a significant detrimental effect on the model.

```r
names(Cigarette)
```

```r
## remove the string data first, and taxs as we already have a tax column via subsetting
CigSub = Cigarette[, 3:8]
```

```r
head(CigSub)
```

```r
FitAll = lm(packpc ~ ., data = CigSub)
```

```r
summary(FitAll)
```

```r
## p-value is most likely infalted here due to no elimation of predictor variables happening yet within the stepwise repgression
```

### Creating stepwise regression through backward elimination

Begin with our model full of predictors, and work barkwards. Eliminating one at a time, until the model is a good fit. We still judge the model based upon its AIC, Akaike Information Criteria. 
This metic yeilds a measure of the relative quality of a model, comparing two models against each other. The model with the smaller AIC is generally accepted to be the better of the two models.

```r
step(FitAll, direction = 'backward')
```

The first iteration has an AIC of 3065.79, with a recommendation to remove the population variable first in order to better streamliine the model to best predict packs per capita. Upon removing the population variable, the model adjusts to have the best possible fit through bakckwar elimination, as the iteration stop after two. 

Stepwise regression through backward elimination yeilds a prdiction model of income, tax, cpi, and avgprs for packs per capita.

```r
## next create that model to get a summary of it
```

```r
fitsome = lm(packpc ~ cpi + income + tax + avgprs, data = CigSub)
```

```r
summary(fitsome)
```

Most of these variables are significant at <.05 with the exception of income, showing significance at .01, with a p-value of .056317 per the model.

```r
## However the adjusted R-squared value of 0.3865 shows that this model only accounts for 38.7% of the variation in the packpc variable
## The F-statistic and its p-value indicate that this model is better than no model at all, p-value <.05
```

### Forward Selection

### Next, try forward selection to develop a best fit model adding IV predictors one at a time

```r
fitstart = lm(packpc ~ 1, data = CigSub)
## start with no predictors in our model. This is the original model of our forward selection process
```

```r
summary(fitstart)
## notice the estimate for the intercept in 106.455, which is simply the the approximate avarage packpc for the Cigarette data set
```

```r
median(CigSub$packpc)
```

Begin forward selcetion using the step() function, and specifcy the scope, ie. which variables we would like to examine for predictor significance. 

```r
## type out every variable, or use the FitAll model that we create early for our backward elimination
step(fitstart, direction = 'forward', scope = (formula(FitAll)))
```

The stepwise regression via forward selection came up with the same model as backward elimination.

```r
fitsome2 = lm(packpc ~ avgprs + cpi + tax + income, data = CigSub)
```

```r
summary(fitsome2)
## same significance levels too
```

### Hybrid Stepwise
#### Forward and Backward Selection

```r
## begin with no predictors like in Forward selection. Specify the direction of both, and the scope.
```

```r
step(fitstart, direction="both", scope=formula(FitAll))
```

```r
fitsome3 = lm(packpc ~ avgprs + cpi + tax + income, data = CigSub)
```

```r
summary(fitsome3)
```

**This model is approximately the best fit model for predicting packs per capita with regression. It accounts for 38.65% of the variation of the dependent variable, packpc.**
